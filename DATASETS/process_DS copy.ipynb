Filtering by SNR:

```python
idx = (df_21['SNR'] > 50) & (df_21['SNR'] < 150)
df1 = df_21[idx]
df1
```

```python
def filter_snr(df):
    snr = df['SNR']
    idx = (50 < snr) & (snr < 150)
    dff = df[idx]
    return dff
```

```python
dff_2sb = filter_snr(df_2sb)
dff_2mb = filter_snr(df_2mb)
dff_gm = filter_snr(df_gm)
dff_21 = filter_snr(df_21)
dff_211 = filter_snr(df_211)
```

```python
dataframes = [dff_2sb, dff_2mb, dff_gm, dff_21, dff_211]
old_dataframes = [df_2sb, df_2mb, df_gm, df_21, df_211]
i = 0
for df in dataframes:
    snr = df['SNR']
    print(min(snr),max(snr), 'Surviving points: ', len(snr), 'of', len(old_dataframes[i]), ', ', 100*(len(snr)/len(old_dataframes[i])), '%')
    i += 1
```

```python
for df in dataframes:
    df.drop('fs',axis=1,inplace=True)
```

```python
dff_21[:5]
```

Saving filtered datasets:

```python
names = ['2ndSCSDM_SingleBit','2ndSCSDM_MultiBit','2ndGmCSDM_V2','3rd21SCSDM','4th211SCSDM']
for i in range(5):
    df = dataframes[i]
    df_name = names[i]
    df_path = str(MYPATH + '\\FinalDataSets\\dataset_' + df_name + '.csv')
    df.to_csv(df_path, index=False)
```

Total dataset for classifiers

```python
specs = ['SNR','Bw','Power']
dfs_2sb = dff_2sb[specs]
dfs_2mb = dff_2mb[specs]
#dfs_gm = dff_gm[specs]
dfs_21 = dff_21[specs]
dfs_211 = dff_211[specs]
```

```python
dfs_21[:5]
```

```python
specsgm = ['SNR','BW','Power']
dfs_gm = dff_gm[specsgm]
```

```python
dfs_gm[:5]
```

```python
dfs_gm = dfs_gm.rename(columns={'BW': 'Bw'})
```

```python
dfs_gm[:5]
```

```python
dfs = [dfs_2sb, dfs_2mb, dfs_gm, dfs_21, dfs_211]
for i in range(len(dfs)):
    df = dfs[i]
    l = len(df)
    cat_col = l*[names[i]]
    df['Category'] = cat_col
```

```python
dfs_gm[:5]
```

```python
min_lengths = min([len(df) for df in dfs])
dflist = len(dfs)*[1]
for i in range(len(dfs)):
    df = dfs[i]
    dflist[i] = df.sample(n = min_lengths)
```

```python
df_total = pd.concat(dflist)
```

```python
df_total
```

Saving data:

```python
df_total_path = str(MYPATH + '\\FinalDataSets\\dataset_total_V2.csv')
df_total.to_csv(df_total_path, index=False)
```

Plotting final result:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(221, projection='3d')

# Create a color palette for the categories
palette = sns.color_palette(n_colors=len(df_total['Category'].unique()))

# Plot each category separately with a legend
for i, cat in enumerate(df_total['Category'].unique()):
    subset = df_total[df_total['Category'] == cat]
    ax.scatter(subset['SNR'], np.log10(subset['Bw']), subset['Power'], label=cat, c=[palette[i]], s=10, alpha=0.5, marker='.')

# Set axis labels
ax.set_xlabel('SNR (dB)')
ax.set_ylabel('Bw (Hz, 10^)')
ax.set_zlabel('Power (W)')

ax2 = fig.add_subplot(224)

for i, cat in enumerate(df_total['Category'].unique()):
    subset = df_total[df_total['Category'] == cat]
    ax2.scatter(subset['SNR'], np.log10(subset['Bw']), label=cat, c=[palette[i]], s=10, alpha=0.5, marker='.')

ax2.set_xlabel('SNR (dB)')
ax2.set_ylabel('BW (Hz, 10^)')

ax3 = fig.add_subplot(223)

for i, cat in enumerate(df_total['Category'].unique()):
    subset = df_total[df_total['Category'] == cat]
    ax3.scatter(subset['Power'], np.log10(subset['Bw']), label=cat, c=[palette[i]], s=10, alpha=0.5, marker='.')

ax3.set_xlabel('Power (W)')
ax3.set_ylabel('BW (Hz, 10^)')

ax4 = fig.add_subplot(222)

for i, cat in enumerate(df_total['Category'].unique()):
    subset = df_total[df_total['Category'] == cat]
    ax4.scatter(subset['SNR'], subset['Power'], label=cat, c=[palette[i]], s=10, alpha=0.5, marker='.')

ax4.set_xlabel('SNR (dB)')
ax4.set_ylabel('Power (W)')


# Add a legend
legend = ax4.legend()
for handle in legend.legend_handles:
    handle.set_sizes([100])


# Save the plot
plt.savefig('total_dataset_V2_proj.png')

# Show the plot
plt.show()
```
